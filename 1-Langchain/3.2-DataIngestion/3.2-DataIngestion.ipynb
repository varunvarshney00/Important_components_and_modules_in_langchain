## 124. Text Splitting Technique-HTML Header Text Splitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<!DOCTYPE html>
<html>
<body>
    <div>
        <h1>Foo</h1>
        <p>Some intro text about Foo.</p>
        <div>
            <h2>Bar main section</h2>
            <p>Some intro text about Bar.</p>
            <h3>Bar subsection 1</h3>
            <p>Some text about the first subtopic of Bar.</p>
            <h3>Bar subsection 2</h3>
            <p>Some text about the second subtopic of Bar.</p>
        </div>
        <div>
            <h2>Baz</h2>
            <p>Some text about Baz</p>
        </div>
        <br>
        <p>Some concluding text about Foo</p>
    </div>
</body>
</html>
"""

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3")
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
html_header_split = html_splitter.split_text(html_string)
html_header_split
```

```python
url = "https://plato.stanford.edu/entries/goedel/"

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
    ("h4", "Header 4"),
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)

html_header_splits = html_splitter.split_text_from_url(url)
html_header_splits
```

## 129. VectorStores-FAISS


```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
loader = TextLoader("speech.txt")
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size = 1000, chunk_overlap = 30)
docs = text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings = OllamaEmbeddings(model="llama3.2:1b")
db = FAISS.from_documents(docs, embeddings)
db
```

```python
### querying
```

```python
query = "How did joe bryant lived his life?"
docs = db.similarity_search(query)
docs
```

```python
docs[0]
```

```python
docs[0].page_content
```

```python
## as a retriever
## we can also convert the vectorstore into a retriever class. this allows us to easily use it in other langchain methods, which largely work with retrievers
```

```python
retriever = db.as_retriever()
docs = retriever.invoke(query)
docs
```

```python
docs[0]
```

```python
docs[0].page_content
```

```python
## similarity search with score
## there are some faiss specific methods. one of them is similarity_search_with_score, which allows you to return not only the documents but also the distance score of the query to them. The returned distance score is L2 distance. therefore, a lower score is better
```

```python
docs_and_score = db.similarity_search_with_score(query)
docs_and_score
```

```python
embedding_vector = embeddings.embed_query(query)
embedding_vector
```

```python
docs_score = db.similarity_search_by_vector(embedding_vector)
docs_score
```

```python
## saving and loading
```

```python
db.save_local("faiss_index")
```

```python
new_db = FAISS.load_local("faiss_index", embeddings, allow_dangerous_deserialization=True)
docs = new_db.similarity_search(query)
docs
```

## 130. VectorStore And Retriever - ChromaDB

```python
## CHROMA
## chroma is a ai-native open-source vector database focused on developer poductivity and happiness. chroma is licensed under apache 2.0
```

```python
## building sample vector db
```

```python
from langchain_chroma import Chroma
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader("speech.txt")
data = loader.load()
data
```

```python
#split
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 0)
splits = text_splitter.split_documents(data)
```

```python
embedding = OllamaEmbeddings(model="llama3.2:1b")
vectordb = Chroma.from_documents(documents = splits, embedding = embedding)
vectordb
```

```python
## query it
```

```python
query = "Was kobe bryant's wife supportive?"
docs = vectordb.similarity_search(query)
docs[0].page_content
```

```python
## saving to the disk
```

```python
vectordb = Chroma.from_documents(documents=splits, embedding=embedding, persist_directory="./chroma_db")
```

```python
# load from disk
db2 = Chroma(persist_directory="./chroma_db", embedding_function=embedding)
docs = db2.similarity_search(query)
print(docs[0].page_content)
```

```python
# similarity search with score
docs = vectordb.similarity_search_with_score(query)
docs
```

```python
# retriever option
retriever = vectordb.as_retriever()
retriever.invoke(query)[0].page_content
```
